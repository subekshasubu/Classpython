```python
# creating csv file

s=""
n = int(input("enter n="))
for i in range(n):
    name=input("enter name=")
    phone=int(input("enter phone="))
    age= int(input("enter age="))
    address= input("enter address=")
    bloodgroup= input("enter blood group=")
    info= f"{name},{age},{address},{phone},{bloodgroup}\n"
    s= s+info
b= open('info.csv','w')
b.write(s)
b.close()
```

```python


import csv
z=[]
n= int(input("ener the value of rows n="))
for i in range(n):
    name=input("enter the name=")
    phone=int(input("enter the phone="))
    age=int(input("enter age="))
    address= input("enter address=")
    info=[i,name,phone,age,address]
    z.append(info)
print(z)
with open('details.csv','a') as f:
    x= csv.writer(f)
    for j in z:
        x.writerow(j)
    
```

```python
#DICT WRITER

import csv
csv_columns=["Name","Address","Age"]
dicty=[{'Name':'subu','address':'ktm','age':23},
     {'Name':'suru','address':'patan','age':24}]
try:
    with open('data.csv','w') as csvfile:
        writer= csv.DictWriter( csvfile , fieldnames=csv_columns)
        writer.writeheader()
        for data in dicty:
            writer.writerow(data)
except IOError:
    print("error occured")
```
